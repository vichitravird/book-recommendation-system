```python
ratings=pd.read_csv(r'C:\Users\vichi\OneDrive - The University of Texas at Dallas\Other Documents\Documents\book-recommendation-system\archive\BX-Book-Ratings.csv', sep=";", on_bad_lines='skip', encoding='latin-1')
```

```python
ratings_with_books=ratings.merge(books, on="ISBN")
```

```python
num_rating=ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
num_rating
```

```python
num_rating.rename(columns={'rating':'num_of_rating'}, inplace=True)
```

```python
num_rating.head()
```

```python
ratings_with_books.head()
```

```python
final_rating=ratings_with_books.merge(num_rating, on='title')
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating=final_rating[final_rating['num_of_rating']>=500]
```

```python
final_rating
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'], inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(5, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse=csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion=model.kneighbors(book_pivot.iloc[7,:].values.reshape(1,-1), n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
def recommend_book(book_name):
    book_id=np.where(book_pivot.index==book_name)[0][0]
    distance, suggestion=model.kneighbors(book_pivot.iloc[7,:].values.reshape(1,-1), n_neighbors=6)
    
    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            print(j)
```
